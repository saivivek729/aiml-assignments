```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
a = pd.read_csv("/content/ml.csv")
numerical_features = a.select_dtypes(include=['int64', 'float64']).columns
categorical_features = a.select_dtypes(include=['object']).columns
numerical_data = a[numerical_features]
categorical_data = a[categorical_features]
scaler = MinMaxScaler()
numerical_scaled = scaler.fit_transform(numerical_data)
numerical_scaled_df = pd.DataFrame(numerical_scaled, columns=numerical_data.columns)
scaled_a = pd.concat([numerical_scaled_df, categorical_data], axis=1)
print(scaled_a.head())
```
